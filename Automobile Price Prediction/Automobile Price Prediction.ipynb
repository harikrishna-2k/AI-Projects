```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
dataset = pd.read_csv('Automobile price data _Raw_.csv')
dataset = dataset[(dataset.values != '?').all(axis=1)]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for i in range(2,8+1):
    dataset.iloc[:,i]=le.fit_transform(dataset.iloc[:,i])
for i in range(14,17+1):
    dataset.iloc[:,i]=le.fit_transform(dataset.iloc[:,i])
dataset.iloc[:,19]=le.fit_transform(dataset.iloc[:,19])
dataset
```

```python
x=dataset.iloc[:, 0:25]
y=dataset.iloc[:,25]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)
print('Accuracy when using Linear regression is', regression_model.score(x, y))
```

```python
import matplotlib.pyplot as plt
plt.plot(x.iloc[:,2],y)
plt.title('Car popularities')
plt.xlabel('Car Models')
plt.ylabel('Prices')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
regression_model = DecisionTreeRegressor()
regression_model.fit(x_train, y_train)
y_pred= regression_model.predict(x_test)
print('Accuracy when using Decision Tree Regressor is',regression_model.score(x, y))

from sklearn import tree
tree.plot_tree(regression_model)
import matplotlib.pyplot as plt
plt.show(le)
```

```python
from sklearn.ensemble import RandomForestRegressor
regression_model = RandomForestRegressor(n_estimators=2000)
regression_model.fit(x_train, y_train)
y_pred= regression_model.predict(x_test)
print('Accuracy when using Random Forest Regression is',regression_model.score(x, y))
```
